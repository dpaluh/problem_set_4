```python
import csv
AmphiBIO=open("/home/dpaluh/problem_set_4/AmphiBIO_v1.csv")


reader = csv.DictReader(AmphiBIO)

Taxonomy_dict={}

# Read through the file and make a dictionary for genus names.
# This gets a unique list of genus names.

for Line in reader:
    if Line['Genus'] not in Taxonomy_dict:
        Taxonomy_dict[Line['Genus']]=[Line['Family'],Line['Order']]   
        
print(Taxonomy_dict)
```

```python
# Add the Taxonomy_dict codes to the Taxonomy table
from sqlalchemy.orm import sessionmaker
conn = engine.connect()

def insert_taxonomy(Genus, Family, Order):
    ins=Taxonomy.insert().values(Genus=Genus,
                                Family=Family,
                                Order=Order)
    result = conn.execute(ins)

for key, value in Taxonomy_dict.items(): 
    insert_taxonomy(key,  value[0], value[1])    
```
